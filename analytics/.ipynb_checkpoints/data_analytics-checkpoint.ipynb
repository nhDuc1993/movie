```python
# Import data
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Load data
pd.set_option('display.max_columns', None)

filepath = r'C:\Users\ADMIN\Desktop\python\project'
df = pd.read_csv(filepath + r'\clean_data\fact_movie.csv', encoding='UTF-8', engine='python')
df
```

```python
# Get budget/revenue columns
budget = df['budget']
revenue = df['revenue']
```

```python
# Visualize
fig, axes = plt.subplots(1, 1, figsize = (12, 12))
plt.scatter(budget, revenue, s=10)
plt.xlabel('budget')
plt.ylabel('revenue')
```

```python
# Linear Regression
model = LinearRegression(fit_intercept=True)
x = budget
y = revenue
X = x[:, np.newaxis]
model.fit(X, y)
yfit = model.predict(X)

print("coef =", model.coef_[0])
print("intercept =", model.intercept_)
```

```python
fig, axes = plt.subplots(1, 1, figsize = (12, 12))
plt.scatter(budget, revenue, s=10)
plt.plot(x, yfit, c ='r')
plt.xlabel('budget')
plt.ylabel('revenue')
```

```python
# Get df_test dataframe

filepath = r'C:\Users\ADMIN\Desktop\python\project'
df_org = pd.read_csv(filepath + '\data\movie_dataset.csv', encoding='UTF-8', engine='python', index_col='index')

df_test  = df[['id', 'title', 'inflated_budget','profitability', 'runtime', 'weighted_rating']]

df_test = pd.merge(df_test, df_org[['id', 'genres']], how='left', left_on='id', right_on='id')
df_test = df_test.iloc[df_test['genres'].dropna().index, :]
df_test['genres'] = df_test['genres'].apply(lambda x: x.split()[0] if x.split()[0] != 'Science' else 'Science Fiction' )
df_test = df_test.drop(['id', 'title'], axis=1)
df_test.dropna(inplace=True)
df_test
```

```python
# Pairplot
sns.pairplot(df_test, hue='genres')
```

```python
# GaussianNB

X = df_test.drop(['genres'], axis=1)
y = df_test['genres']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=1)

model = GaussianNB()

model.fit(Xtrain, ytrain)

y_model = model.predict(Xtest)

accuracy_score(ytest, y_model)
```

```python
# Predict if the movie is profitable or not
# Pre-process dataset

filepath = r'C:\Users\ADMIN\Desktop\python\project'
df_genres = pd.read_csv(filepath + '\clean_data\dim_genres.csv', encoding='UTF-8', engine='python')
df_genres = df_genres.groupby('movie_id').agg({'genres': lambda x: ','.join(x)}).reset_index()

df_test  = df[['id', 'inflated_budget','profitability', 'runtime', 'weighted_rating']]

df_test = pd.merge(df_test, df_genres, how='left', left_on='id', right_on='movie_id')
df_test = df_test.iloc[df_test['genres'].dropna().index, :]
df_test = df_test.drop(['movie_id'], axis=1)
df_test.dropna(inplace=True)
df_test
```

```python
# Profit?
df_test['profitability'] = df_test['profitability'].apply(lambda x: 1 if x > 100 else 0)
df_test
```

```python
# Inflated_budget

df_test = df_test[df_test['inflated_budget'] != df_test['inflated_budget'].max()]
g = sns.FacetGrid(df_test, col='profitability', height=10)
g.map(plt.hist, 'inflated_budget', bins=20)
```

```python
# Create bin

df_test['inflated_budget_bin'] = pd.cut(df_test['inflated_budget'], 5)
df_test[['inflated_budget_bin', 'profitability']].groupby(['inflated_budget_bin'], as_index=False).mean()
```

```python
# Apply bins

df_test.loc[ df_test['inflated_budget'] <= 89922109.4, 'inflated_budget'] = 0
df_test.loc[(df_test['inflated_budget'] > 89922109.4) & (df_test['inflated_budget'] <= 179830094.8), 'inflated_budget'] = 1
df_test.loc[(df_test['inflated_budget'] > 179830094.8) & (df_test['inflated_budget'] <= 269738080.2), 'inflated_budget'] = 2
df_test.loc[(df_test['inflated_budget'] > 269738080.2) & (df_test['inflated_budget'] <= 359646065.6), 'inflated_budget'] = 3
df_test.loc[ df_test['inflated_budget'] > 359646065.6, 'inflated_budget'] = 4
df_test.drop('inflated_budget_bin', axis=1, inplace=True)
df_test
```

```python
# Runtime
df_test = df_test[df_test['runtime'] != 0]

g = sns.FacetGrid(df_test, col='profitability', height=10)
g.map(plt.hist, 'runtime', bins=4)
```

```python
df_test['runtime_bin'] = pd.cut(df_test['runtime'], 4)
df_test[['runtime_bin', 'profitability']].groupby(['runtime_bin'], as_index=False).mean()
```

```python
df_test.loc[ df_test['runtime'] <= 115.25, 'runtime'] = 0
df_test.loc[(df_test['runtime'] > 115.25) & (df_test['runtime'] <= 189.5), 'runtime'] = 1
df_test.loc[(df_test['runtime'] > 189.5) & (df_test['runtime'] <= 263.75), 'runtime'] = 2
df_test.loc[ df_test['runtime'] > 263.75, 'runtime'] = 3
df_test.drop('runtime_bin', axis=1, inplace=True)
df_test['runtime'] = df_test['runtime'].astype(int)
df_test
```

```python
# Weighted rating
g = sns.FacetGrid(df_test, col='profitability', height=10)
g.map(plt.hist, 'weighted_rating', bins=4)
```

```python
df_test['weighted_rating_bin'] = pd.cut(df_test['weighted_rating'], 4)
df_test[['weighted_rating_bin', 'profitability']].groupby(['weighted_rating_bin'], as_index=False).mean()
```

```python
df_test.loc[ df_test['weighted_rating'] <= 4.6, 'weighted_rating'] = 0
df_test.loc[(df_test['weighted_rating'] > 4.6) & (df_test['weighted_rating'] <= 5.896), 'weighted_rating'] = 1
df_test.loc[(df_test['weighted_rating'] > 5.896) & (df_test['weighted_rating'] <= 7.193), 'weighted_rating'] = 2
df_test.loc[ df_test['weighted_rating'] > 7.193, 'weighted_rating'] = 3
df_test.drop('weighted_rating_bin', axis=1, inplace=True)
df_test['weighted_rating'] = df_test['weighted_rating'].astype(int)
df_test
```

```python
# Genres
filepath = r'C:\Users\ADMIN\Desktop\python\project'
df_genres = pd.read_csv(filepath + '\clean_data\dim_genres.csv', encoding='UTF-8', engine='python')
pd.merge(df_test[['id','profitability']], df_genres, how='left', left_on='id', right_on='movie_id')
df_genres = pd.merge(df_test[['id','profitability']], df_genres, how='left', left_on='id', right_on='movie_id')
df_genres = df_genres.groupby('genres').mean().profitability.sort_values(ascending=False)
df_genres
```

```python
# Define genre_cal function

def genre_cal(genre):
    s = 0
    for i in genre.split(","):
        s += df_genres[i]
        
    return s/len(genre.split(","))
df_test['genres'] = df_test['genres'].apply(lambda x: genre_cal(x))
```

```python
g = sns.FacetGrid(df_test, col='profitability', height=10)
g.map(plt.hist, 'genres', bins=4)
```

```python
df_test['genres_bin'] = pd.cut(df_test['genres'], 4)
df_test[['genres_bin', 'profitability']].groupby(['genres_bin'], as_index=False).mean()
```

```python
df_test.loc[ df_test['genres'] <= 0.554, 'genres'] = 0
df_test.loc[(df_test['genres'] > 0.554) & (df_test['genres'] <= 0.625), 'genres'] = 1
df_test.loc[(df_test['genres'] > 0.625) & (df_test['genres'] <= 0.696), 'genres'] = 2
df_test.loc[ df_test['genres'] > 0.696, 'genres'] = 3
df_test.drop('genres_bin', axis=1, inplace=True)
df_test['genres'] = df_test['genres'].astype(int)
df_test
```

```python
X = df_test.drop(['id', 'profitability'], axis=1)
y = df_test['profitability']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=1)
```

```python
# Logistic Regression

logreg = LogisticRegression()
logreg.fit(Xtrain, ytrain)
Y_pred = logreg.predict(Xtest)
log_score = round(logreg.score(Xtrain, ytrain) * 100, 2)
log_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(log_score)
print(log_acc)
```

```python
coeff_df = pd.DataFrame(df_test.columns.delete([0,2]))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
# Support Vector Machines

svc = SVC()
svc.fit(Xtrain, ytrain)
Y_pred = svc.predict(Xtest)
svc_score = round(svc.score(Xtrain, ytrain) * 100, 2)
svc_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(svc_score)
print(svc_acc)
```

```python
# Kneighbor
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(Xtrain, ytrain)
Y_pred = knn.predict(Xtest)
knn_score = round(knn.score(Xtrain, ytrain) * 100, 2)
knn_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(knn_score)
print(knn_acc)
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()
gaussian.fit(Xtrain, ytrain)
Y_pred = gaussian.predict(Xtest)
gaussian_score = round(gaussian.score(Xtrain, ytrain) * 100, 2)
gaussian_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(gaussian_score)
print(gaussian_acc)
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(Xtrain, ytrain)
Y_pred = perceptron.predict(Xtest)
perceptron_score = round(perceptron.score(Xtrain, ytrain) * 100, 2)
perceptron_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(perceptron_score)
print(perceptron_acc)
```

```python
# Linear SVC

linear_svc = LinearSVC()
linear_svc.fit(Xtrain, ytrain)
Y_pred = linear_svc.predict(Xtest)
linear_svc_score = round(linear_svc.score(Xtrain, ytrain) * 100, 2)
linear_svc_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(linear_svc_score)
print(linear_svc_acc)
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(Xtrain, ytrain)
Y_pred = decision_tree.predict(Xtest)
decision_tree_score = round(decision_tree.score(Xtrain, ytrain) * 100, 2)
decision_tree_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(decision_tree_score)
print(decision_tree_acc)
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(Xtrain, ytrain)
Y_pred = random_forest.predict(Xtest)
random_forest_score = round(random_forest.score(Xtrain, ytrain) * 100, 2)
random_forest_acc = round(accuracy_score(ytest, Y_pred) * 100, 2) 
print(random_forest_score)
print(random_forest_acc)
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Linear SVC', 'Decision Tree'],
    'Score': [svc_score, knn_score, log_score, 
              random_forest_score, gaussian_score, perceptron_score, 
              linear_svc_score, decision_tree_score],
    'Accuracy': [svc_acc, knn_acc, log_acc, 
              random_forest_acc, gaussian_acc, perceptron_acc, 
              linear_svc_acc, decision_tree_acc]})
models
```
