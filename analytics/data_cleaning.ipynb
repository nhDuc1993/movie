```python
# Import libraries
import pandas as pd
import re
```

```python
# Load dataset
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)

filepath = r'C:\Users\ADMIN\Desktop\python\project'
df = pd.read_csv(filepath + r'\data\movie_dataset.csv', encoding='UTF-8', engine='python', index_col='index')
df
```

```python
# Check overall information:
print(df.info())
print("-" * 50)
print(df.describe())
```

```python
# Create 'day', 'month', 'year' columns
df['day'] = df['release_date'].apply(lambda x: int(x.split('-')[-1]) if type(x) != float else 0) 
df['month'] = df['release_date'].apply(lambda x: int(x.split('-')[1]) if type(x) != float else 0) 
df['year'] = df['release_date'].apply(lambda x: int(x.split('-')[0]) if type(x) != float else 0)
df[['id', 'title', 'day', 'month', 'year']]
```

```python
# Get columns with encoded unicode characters
df_dtypes = pd.DataFrame(df.dtypes)
col_with_u = []

for i in df_dtypes[df_dtypes[0] == 'object'].index:
    for j in df[i].dropna():
        if '\\u' in j:
            col_with_u.append(i)
            break

col_with_u
```

```python
# Decode unicode characters:
for i in ['keywords','cast','director']:
    df[i] = df[i].apply(lambda x: x.encode().decode("unicode-escape") if type(x) != float else x)
df[['keywords','cast','director']]
```

```python
# Missing budget/revenue values
df.loc[(df['budget'] == 0) | (df['revenue'] == 0), ['id', 'title', 'budget','revenue']]
```

```python
# Append missing budget/revenue information
df_the_numbers = pd.read_csv(filepath + r'\budget_revenue_crawl\the_numbers_crawl_res\the_numbers_crawl.csv')
df_the_numbers.set_index('id', inplace=True)

df_google = pd.read_csv(filepath + r'\budget_revenue_crawl\google_crawl_res\google_crawl.csv')
df_google.set_index('id', inplace=True)

df.set_index('id', inplace=True)

for i in df_google.index:
    if df_google.loc[i, 'budget'] > df_the_numbers.loc[i, 'budget']:
        df_the_numbers.loc[i, 'budget'] = df_google.loc[i, 'budget']
    if df_google.loc[i, 'revenue'] > df_the_numbers.loc[i, 'revenue']:
        df_the_numbers.loc[i, 'revenue'] = df_google.loc[i, 'revenue']

for i in df_the_numbers.index:
    if df_the_numbers.loc[i, 'budget'] > df.loc[i, 'budget']:
        df.loc[i, 'budget'] = df_the_numbers.loc[i, 'budget']
    if df_the_numbers.loc[i, 'revenue'] > df.loc[i, 'revenue']:
        df.loc[i, 'revenue'] = df_the_numbers.loc[i, 'revenue']

df.reset_index(inplace=True)
```

```python
# Clean budget/revenue outliers/errors)
df.loc[4267, 'budget'] = 1378000
df.loc[4267, 'revenue'] = 3900000
df.loc[3693, 'budget'] = 9000000
df.loc[3693, 'revenue'] = 728 

df = df[df['budget'] >= 10000]
df = df[df['revenue'] >= 10000]
df.reset_index(inplace=True, drop=True)

df[['id', 'title', 'budget','revenue']]
```

```python
# Load inflation dataset:
df_inflation = pd.read_csv(filepath + r'\data\inflation_dataset.csv')

# Define 2021 inflation_cal function:
def inflation_cal(number, year):
    if year < 1920:
        year = 1920
    if year == 2021:
        return number
    
    idx = df_inflation[df_inflation['year'] == year].index[0]
    for i in df_inflation.index[idx+1:]:
        number = number * (1+df_inflation.loc[i, 'inflation rate'])
        
    return int(number)

# Create inflation columns:
df['inflated_budget'] = df[['budget', 'year']].apply(lambda x: inflation_cal(x['budget'], x['year']), axis=1)
df['inflated_revenue'] = df[['revenue', 'year']].apply(lambda x: inflation_cal(x['revenue'], x['year']), axis=1)

df[['id', 'title', 'budget','revenue', 'inflated_budget', 'inflated_revenue']]
```

```python
# Create profitability column
def profitability(budget, revenue):
    p = revenue/budget * 100
    return round(p,2)

df['profitability'] = df[['budget', 'revenue']].apply(lambda x: profitability(x.budget, x.revenue), axis=1)
df[['id', 'title', 'budget','revenue', 'profitability']]
```

```python
# Create df_gernes table
df_genres = df[['id', 'genres']]
df_genres.dropna(inplace = True)

# Define gerne-splitting function:
def gernes_split(genres):
    genres_split = genres.split(" ")

    for i in range(len(genres_split)-1):
        if ' '.join([genres_split[i], genres_split[i+1]]) == 'Science Fiction':
            genres_split.remove('Science')
            genres_split.remove('Fiction')
            genres_split.append('Science Fiction')
            break
    
    return genres_split

# Split values into lists:
df_genres['genres'] = df_genres['genres'].apply(lambda x: gernes_split(x))

# Create dim_gernes table:
dim_genres = pd.DataFrame(columns = ['movie_id', 'genres'])

for i in df_genres.index:
    temp_dim_genres = pd.DataFrame(columns = ['movie_id', 'genres'])
    temp_dim_genres['genres'] = df_genres.loc[i, 'genres']
    temp_dim_genres['movie_id'] = df_genres.loc[i, 'id']
    dim_genres = dim_genres.append(temp_dim_genres, ignore_index = True)
    
# Remove TV Movie
dim_genres = dim_genres[~((dim_genres['genres'] == "TV") | (dim_genres['genres'] == "Movie"))]

dim_genres
```

```python
# Create df_prod_comp table:
df_prod_comp = df[['id', 'production_companies']]

# Create dim_prod_comp table:
dim_prod_comp = pd.DataFrame(columns = ['name', 'id', 'movie_id'])

for i in df_prod_comp.index:
    temp_prod_comp = pd.read_json(df_prod_comp.iloc[i,1])
    temp_prod_comp['movie_id'] = df_prod_comp.iloc[i,0]
    dim_prod_comp = dim_prod_comp.append(temp_prod_comp, ignore_index = True)

# Rename and rearrange columns:
dim_prod_comp.rename(columns = {'name' : 'comp_name', 'id' : 'comp_id'}, inplace = True)
dim_prod_comp = dim_prod_comp.reindex(columns = ['movie_id', 'comp_id', 'comp_name'])

dim_prod_comp
```

```python
# Create df_prod_ctry table:
df_prod_ctry = df[['id', 'production_countries']]

# Create dim_prod_ctry table:
dim_prod_ctry = pd.DataFrame(columns = ['iso_3166_1', 'name', 'movie_id'])

for i in df_prod_ctry.index:
    temp_prod_ctry = pd.read_json(df_prod_ctry.iloc[i,1])
    temp_prod_ctry['movie_id'] = df_prod_ctry.iloc[i,0]
    dim_prod_ctry = dim_prod_ctry.append(temp_prod_ctry, ignore_index = True)

# Rename and rearrange columns:
dim_prod_ctry.rename(columns = {'name' : 'ctry_name', 'iso_3166_1' : 'ctry_code'}, inplace = True)
dim_prod_ctry = dim_prod_ctry.reindex(columns = ['movie_id', 'ctry_code', 'ctry_name'])

dim_prod_ctry
```

```python
# Create df_lang table:
df_lang = df[['id', 'original_language', 'spoken_languages']]

# Create dim_lang table:
dim_lang = pd.DataFrame(columns = ['iso_639_1', 'name', 'movie_id', 'org_lang'])

for i in df_lang.index:
    temp_lang = pd.read_json(df_lang.iloc[i,2])
    temp_lang['movie_id'] = df_lang.iloc[i,0]
    temp_lang['org_lang'] = df_lang.iloc[i,1]
    dim_lang = dim_lang.append(temp_lang, ignore_index = True)

# Get ISO-639 table from Wikipedia   
wiki_url = 'https://en.wikipedia.org/wiki/List_of_ISO_639-1_codes'
df_wiki = pd.read_html(wiki_url)[1][['ISO language name', '639-1']]
dct_wiki = df_wiki.set_index('639-1').to_dict()['ISO language name']

# Fill in the missing information:
for i in dim_lang.index:
    if dim_lang.loc[i, 'name'] == '':
        try:
            dim_lang.loc[i, 'name'] = dct_wiki[dim_lang.loc[i, 'iso_639_1']]
        except:
            pass

dim_lang_right = dim_lang[['iso_639_1', 'name']].drop_duplicates()
dim_lang = pd.merge(dim_lang, dim_lang_right, how='left', left_on='org_lang', right_on='iso_639_1')

dim_lang.loc[dim_lang['iso_639_1_x'] == 'sh', 'name_x'] = 'Serbo-Croatian'
dim_lang.loc[dim_lang['org_lang'] == 'nb', 'name_y'] = 'Norsk'

dim_lang.drop(['iso_639_1_y'], axis=1, inplace=True)

# Rename and rearrange columns:
dim_lang.rename(columns = {'iso_639_1_x' : 'spk_lang', 'name_y' : 'org_lang_name', 'name_x' : 'spk_lang_name'}, inplace = True)
dim_lang = dim_lang.reindex(columns = ['movie_id', 'org_lang','org_lang_name', 'spk_lang', 'spk_lang_name'])

dim_lang
```

```python
# Create df_crew table:
df_crew = df[['id', 'crew']]

# Define json_clean functions:
def json_clean(json):
    r = re.findall(r'"(.*?)"', json)
    for i in r:
        c = i.replace('\'', '')
        json = json.replace(i , c)
    
    r = re.findall(r"'(.*?)'", json)
    for i in r:
        c = i.replace('"', '')
        json = json.replace(i , c)    
        
    json = json.replace('\'', '"')
    
    return pd.read_json(json).rename(columns={'name': 'crew_name', 'id' : 'crew_id' })

# Create the dim_crew table:
dim_crew = pd.DataFrame(columns = ['crew_name', 'gender', 'department', 'job', 'credit_id', 'crew_id', 'movie_id'])

for i in df_crew.index:
    temp_crew = json_clean(df_crew.iloc[i,1])
    temp_crew['movie_id'] = df_crew.iloc[i,0]
    dim_crew = dim_crew.append(temp_crew, ignore_index = True)
    
dim_crew['crew_name'] = dim_crew['crew_name'].apply(lambda x: x.encode().decode("unicode-escape") if type(x) != float else x)

# Rename and rearrange columns:
dim_crew = dim_crew.reindex(columns = ['movie_id', 'crew_name','gender', 'credit_id', 'department', 'job'])

dim_crew
```

```python
# Drop columns
df.drop(['genres', 'original_language', 'production_companies', 'production_countries', 'spoken_languages', 'crew', 'status'], axis = 1, inplace=True)
df
```

```python
# Create weighted rating column:
## R: average rating of the movie
## v: the amount of votes
## m: the mininum amount of votes to be ranked in the chart
## C: mean rating of the whole set

def weighted_rating(R, v, m, C):
    if v == 0:
        return None
    
    WR = (v / (v+m)) * R + (m / (v+m)) * C 
    return round(WR,3)

m = 54 
C = sum(df['vote_average'] * df['vote_count']) / sum(df['vote_count'])

df['weighted_rating'] = df[['vote_average', 'vote_count']].apply(lambda x: weighted_rating(x['vote_average'], x['vote_count'], m, C), axis=1)

df[['id', 'title', 'vote_count', 'vote_average', 'weighted_rating']]
```

```python
# Create correlation table
df_corr = abs(df[['budget', 'revenue', 'runtime', 'profitability', 'weighted_rating', 'popularity']].corr())
df_corr.reset_index(inplace=True)

df_corr
```
